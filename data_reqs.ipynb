```python
import cdsapi, pathlib, time, concurrent.futures

# ---- edit your bounding box here ----


OUTDIR = pathlib.Path("/Volumes/Dhruv_External_Disk/climate_data/london")
OUTDIR.mkdir(parents=True, exist_ok=True)

```

```python
def download_daily_mean(year):
    dataset = "sis-european-risk-extreme-precipitation-indicators"
    request = {
        "spatial_coverage": ["city"],
        "variable": ["standardised_precipitation_exceeding_fixed_percentiles"],
        "city": ["london"],
        "product_type": ["era5_2km"],
        "temporal_aggregation": ["daily"],
        "percentile": ["95th"],
        "period": [str(year)]
    }
    client = cdsapi.Client()
    target = OUTDIR / f"ldn_ex_tp_{year}.zip"
    if target.exists():
        print(f"✓ {target.name} exists - skip"); return
    
    client.retrieve(dataset, request=request, target=str(target))

from tqdm.auto import tqdm

for yr in tqdm(range(2000, 2019), desc="Downloading years"): 
    download_daily_mean(yr)
```
